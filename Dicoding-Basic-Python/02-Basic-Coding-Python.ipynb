# Basic Coding with Python

## Data Type

### Float & int

```python
a = 5
print(a, "is of type", type(a))
a = 2.0
print(a, "is of type", type(a))
a = 1+2j
print(a, "is complex number?", isinstance(1+2j,complex))
```

```python
# tidak ada pembatasan panjang int seperti dalam C/C++

x=[0]*10005;              #inisialisasi array 0 sebanyak 10005; x[0]=0
x[1]=1;                   #x[1]=1
 
for j in range(2,10001):
      x[j]=x[j-1]+x[j-2]  # Fibonacci
print(x[10000])
```

### String

```python
s = "Hello World!"
print(s[4])
print(s[6:11])
s[5]="d"                  # elemen dalam string tidak immutable
```

### List/Array

```python
x = [5,10,15,20,25,30,35,40]
print(x[5])
print(x[-1])
print(x[3:5])
print(x[:5])
print(x[-3:])
print(x[1:7:2])

# dapat di imitasi
x = [1,2,3]
x[2]=4
print(x)
x.append(5)
print(x)

spam = ['cat', 'bat', 'rat', 'elephant']
del spam[2]
print(spam)
```

```python
t = (5,'program', 1+3j)
t[1]
t[0:3]
t[0]=10
```

### Set => kumpulan item bersifat unik dan tanpa urutan (unordered collection)

```python
a = {1,2,2,3,3,3}
print(a)

a = {1,2,3}
print(a[1])
```

### Dictionary =>  kumpulan pasangan kunci-nilai (pair of key-value) yang bersifat tidak berurutan

```python
d = {1:'value','key':2}
print(type(d))
print("d[1] = ", d[1]);
print("d['key'] = ", d['key']);
 
# Generates error, must call it using the key
print("d[2] = ", d[2]);
```

## Konversi (conversion, cast) antar tipe data

```python
print(float(5))
print(int(10.6))
print(int(-10.6))
print(float('2.5'))
print(str(25))
print(set([1,2,3]))
print(tuple({5,6,7}))
print(list('hello'))
print(dict([[1,2],[3,4]]))
```

## I/O in Python

```python
name = "John"
age = 15
print("%s is %d years old." % (name, age))

# secara otomatis akan terkonversi
mylist = [1,2,3]
print("A list: %s" % mylist)

num = input('Enter a number: ')
print(num)

# menggunakan eval untuk mengkonversi string berisi ekspresi matematika
print(eval('2+3'))
```

## Duck Typing
- Sebuah konsep, tipe atau kelas dari sebuah obyek tidak lebih penting daripada metode yang menjadi perilakunya.
- Hanya memberikan keleluasaan pada developer untuk tidak perlu mencemaskan tentang tipe atau kelas dari sebuah obyek
- Untuk memeriksa dan mengetahui tipe sebuah obyek, Anda cukup memastikan metode/fungsi/behavior dari obyek tersebut
- Misalnya fungsi len() untuk mengetahui panjang string, yang tidak berlaku pada variabel numerik (misalnya integer)

```python
x=333
len(x) # will raise an error, because len is for string
```
