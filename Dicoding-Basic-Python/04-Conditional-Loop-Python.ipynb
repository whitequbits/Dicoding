## Basic Conditional Expression

````python
amount = int(input("Enter amount: "))
if amount<1000:
   discount = amount*0.05
   print ("Discount",discount)
elif amount<5000:
   discount = amount*0.10
   print ("Discount",discount)
else:
   discount = amount*0.15
   print ("Discount",discount)
print ("Net payable:",amount-discount)
````

## Looping Expressions
### For Loop

````python
for letter in 'Python':  # First Example
    print('Current Letter: {}'.format(letter))
fruits = ['banana', 'apple', 'mango']
for fruit in fruits:  # Second Example
    print('Current fruit: {}'.format(fruit))
````

````python
fruits = ['banana', 'apple', 'mango']
for index in range(len(fruits)):
    print('Current fruit: {}'.format(fruits[index]))
````

### While Loop

````python
count = 0
while (count < 5):
    print('The count is: {}'.format(count))
    count = count + 1
````

````python
# beaware of infinite loop

var = 1
while var == 1:  # This constructs an infinite loop
    num = input('Enter a number: ')
    print('You entered: {}'.format(num))
 
 
while True:  # This constructs an infinite loop
    num = input('Enter a number: ')
    print('You entered: {}'.format(num))
````

````python
# short while loop

while (var1): do_something()
````

````python
for i in range(0, 5):
    for j in range(0, 5 - i):
        print('*', end='')
    print()
````

## Break Argument
break and exit the loop

````python

for letter in 'Python':     # First Example
    if letter == 'h':
        break
    print('Current letter: {}'.format(letter))
````

````python
var = 10                    # Second Example
while var > 0:             
    print('Current variable value: {}'.format(var))
    var = var - 1
    if var == 5:
        break
````

## Continue Argument
break but continue to the next loop

````python
for letter in 'Python':     # First Example
    if letter == 'h':
        continue
    print('Current letter: {}'.format(letter))
````

````python
for a in [0, 1, -1, 2, -2, 3, -3]:     #Second Example
    if a <= 0:
        continue
    print('Elemen positif: {}'.format(a))
````

````python
var = 10                    # Third Example
while var > 0:             
    var = var -1
    if var == 5:
        continue
    print('Current variable value: {}'.format(var))
````

## Else after For Loop

````python
for item in container:
    if search_something(item):
        # Found it!
        process(item)
        break
else:
    # Didn't find anything..
    not_found_in_container()
````

````python
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print( n, 'equals', x, '*', n/x)
            break
    else:
        # loop fell through without finding a factor
        print(n, 'is a prime number')
````

## Else after While Loop

````python
n = 5
while n > 0:
    n = n - 1
    if n == 2:
        break
    print(n)
else:
    print("Loop is finished")
````

## Pass Argument

````python
for letter in 'Python':
    if letter == 'h':
        pass
        print ("This is pass block")
    print('Current letter: {}'.format(letter))
````

## Handling with Exception Error

````python
import sys
data=''
while(data!='exit'):
    try:
        data=input('Please enter an integer (type exit to exit): ')
        print('got integer: {}'.format(int(data)))
    except:
        if data == 'exit':
            pass  # exit gracefully without prompt any error
        else:
            print('error: {}'.format(sys.exc_info()[0]))
````

## List Comprehension

```python
new_list = [expression for_loop_one_or_more conditions]
```

````python
#Cara 1
numbers = [1, 2, 3, 4]
squares = []
for n in numbers:
  squares.append(n**2)
print(squares)
````

````python
#Cara 2 List Comprehension
numbers = [1, 2, 3, 4]
squares = [n**2 for n in numbers]
print(squares)
````

````python
list_a = range(1, 10, 2)
x = [[a**2, a**3] for a in list_a]
print(x)
````
