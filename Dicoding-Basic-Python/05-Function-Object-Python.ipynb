### Base Class

````python
class Kalkulator:
    def add(self, a, b):
        return a+b

    def substract(self, a, b):
        return a-b

    def multiply(self, a, b):
        return a*b
    
    def division(self, a, b):
        return a/b

calculator = Kalkulator()
print(calculator.add(2,4))
print(calculator.substract(2,4))
print(calculator.multiply(2,4))
print(calculator.division(2,4))
````

### Derived Class

````python
class KalkulatorSuhu(Kalkulator):
    def __init__(self):
        super().__init__()
        self.temp_celcius = 27
        self.temp_kelvin = 300
        #private variable
        self.__conversion_factor = 273
    
    def set_temp_celcius(self, temp):
        self.temp_celcius = temp
    
    def set_temp_kelvin(self, temp):
        self.temp_kelvin = temp

    def convert_celcius_to_kelvin(self):
        return self.add(self.temp_celcius, self.__conversion_factor)
    
    def convert_kelvin_to_celcius(self):
        return self.substract(self.temp_kelvin, self.__conversion_factor)

    

calculator_suhu = KalkulatorSuhu()
print(calculator_suhu.convert_celcius_to_kelvin())
print(calculator_suhu.convert_kelvin_to_celcius())
````

### Variable Length Argument

````python
def printinfo(*args, **kwargs):
    for a in args:
        print('argumen posisi {}'.format(a))
    for key, value in kwargs.items():
        print('argument kata kunci {}:{}'.format(key, value))
 
 
# Panggil printinfo
printinfo()
printinfo(1, 2, 3)
printinfo(i=7, j=8, k=9)
printinfo(1, 2, j=8, k=9)
printinfo(*(2, 3), **{'i':7, 'j':8})
````

### Global vs Local Variable

````python
# Bad implementation
total = 0  # ini variabel global
def sum(arg1, arg2):
    # tambahkan kedua parameter dan return
    total = arg1 + arg2
    print('Inside the function: {}'.format(total))
    return total
# panggil sum
sum(10, 20)
print('Outside the function: {}'.format(total)) 
````

````python
# Good Implementation
total = 0  # This is global variable.
# fungsi didefinisikan di bawah ini
def sum(arg1, arg2):
    global total # this one will get the total value from local variable
    total = arg1 + arg2 # variabel lokal total
    print('Inside the function local total: ', total)
    return total
 
sum(10, 20)
print('Outside the function (global) total: ', total)
````

## Class / Object

````python
class Kalkulator:
    def add(a,b):
        return a+b

    def substract(a,b):
        return a-b

    def multiply(a,b):
        return a*b

    def division(a,b):
        return a/b

calculator = Kalkulator
print(calculator.add(4,2))
print(calculator.substract(4,2))
print(calculator.multiply(4,2))
print(calculator.division(4,2))
````

### Inheritance

````python
class KalkulatorSuhu(Kalkulator):
    def __init__(self):
        self.C_TO_KELVIN = 273
        super().__init__()

    def celcius_to_kelvin(temp_in_celcius):
        return self.add(temp_in_celcius, C_TO_KELVIN)

    def kelvin_to_celcius(temp_in_kelvin):
        return self.substract(temp_in_kelvin, C_TO_KELVIN)

temperature_calculator = KalkulatorSuhu
print(temperature_calculator.celcius_to_kelvin(30))
````
