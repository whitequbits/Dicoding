```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
from tensorflow import keras
from keras.utils import plot_model
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

### Import Training data

```python
df = pd.read_csv('citrus.csv')
```

```python
df.info()
```

```python
df.head(5)
```

### Preprocessing

#### Replace Label name into 0 and 1

```python
df.name[df.name == 'orange'] = 0.0
df.name[df.name == 'grapefruit'] = 1.0
```

#### Change the data type into array

```python
dataset = df.values
```

#### Assign Feature and Lable

```python
X = dataset[:,1:6] # nx5 array
```

```python
y = dataset[:,0]
```

#### Normalization

```python
# Normalization
min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)
X_scale
```

#### Training Testing Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scale, y, test_size=0.3)
```

To Avoid Error ValueError: Failed to convert a NumPy array to a Tensor (Unsupported object type float)

```python
X_train, X_test, Y_train, Y_test = X_train.astype('float64'), X_test.astype('float64'), Y_train.astype('float64'), Y_test.astype('float64')
```

### Create the model

1. 2 Layer with 32 unit
2. input shape = nx5 array
3. Sigmoid Function to fit the data between 0 and 1

```python
model = tf.keras.Sequential([    
                    keras.layers.Dense(32, activation='relu', input_shape=(5,)),    
                    keras.layers.Dense(32, activation='relu'),    
                    keras.layers.Dense(1, activation='sigmoid'),])
```

##### Create the loss and optimizer function

1. stochastic gradient descent (sgd)
2. binary cross entropy as loss function
3. metrics to benchmark is accuracy

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
plot_model(model, to_file='03-citrus-classification.png')
```

### Fit the model

```python
model.fit(X_train, Y_train, epochs=100)
```

### Test Model

```python
model.evaluate(X_test, Y_test)
```
