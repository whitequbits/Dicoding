```python
import tensorflow as tf
import pandas as pd
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from keras import callbacks
from keras.utils import plot_model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

### Import Dataset

```python
df = pd.read_csv('./datasets/imdb_indonesian_movies_2.csv')
```

```python
df = df.drop(columns=['judul_film'])
```

```python
df.info()
```

```python
df.tail()
```

### Preprocessing

```python
category = pd.get_dummies(df.genre)
```

```python
df_baru = pd.concat([df, category], axis=1)
```

```python
df_baru = df_baru.drop(columns='genre')
```

```python
sentence = df_baru['ringkasan_sinopsis'].values
label = df_baru[['Drama', 'Horor', 'Komedi', 'Laga', 'Romantis']].values
train_sentence, test_sentence, train_label, test_label = train_test_split(sentence, label, test_size=0.2)
```

### Create Tokenizer

```python
tokenizer = Tokenizer(num_words=5000, oov_token='x')
```

### Fit Sentence

```python
tokenizer.fit_on_texts(train_sentence) 
tokenizer.fit_on_texts(test_sentence)
```

### Convert Text to Index

```python
train_sequences = tokenizer.texts_to_sequences(train_sentence)
test_sequences = tokenizer.texts_to_sequences(test_sentence)
```

```python
print(tokenizer.word_index)
```

### Padding

```python
train_padded = pad_sequences(train_sequences) 
test_padded = pad_sequences(test_sequences)
```

### Create Model

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=5000, output_dim=16),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(5, activation='softmax')
])
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
plot_model(model, to_file='09-IMDB-classification.png')
```

### Create Callbacks

```python
class myCallback(callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.9):
      print("\nYour Accuracy >90%!")
      self.model.stop_training = True
```

```python
callbacks = myCallback()
```

### Fit Model

```python
history = model.fit(train_padded, train_label, epochs=50, 
                    validation_data=(test_padded, test_label), verbose=2, callbacks=[callbacks])
```

### Plotting

#### Plot Model Loss

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss Model')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

#### Plot Model Accuracy

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Akurasi Model')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
