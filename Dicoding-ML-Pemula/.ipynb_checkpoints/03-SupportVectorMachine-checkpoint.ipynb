# Support Vector Machine

- sebuah model ML multifungsi yang dapat digunakan untuk klasifikasi linier dan non-linier, regresi, dan bahkan pada pendeteksian outlier
- salah satu metode yang paling populer dalam machine learning.

## Support Vector Machine Classifier

- SVM bekerja dengan membuat decision boundary atau sebuah bidang yang mampu memisahkan dua buah kelas

### SVM Klasifikasi linear

- SVM untuk klasifikasi linear mencari support vector pada setiap kelas.
- SVM mencari margin terbesar atau jalan terlebar dari support vector yang mampu memisahkan kedua kelas.

![SVM Linear](https://dicodingacademy.blob.core.windows.net/academies/202004302114418a9f0636f38334dcfd216f220b238a82.png "SVM Linear")

### SVM Klasifikasi Non linear

- Support Vector Classifier menggunakan sebuah metode yaitu “kernel trick” untuk data non linear
- kernel trick adalah sebuah metode untuk mengubah data pada dimensi tertentu ke dalam dimensi yang lebih tinggi agar dapat dipisahkan secara linier

![](https://dicodingacademy.blob.core.windows.net/academies/202004302120524b2e946499ea49ac5e77a0419e2252b5.png "From 1-D")
![](https://dicodingacademy.blob.core.windows.net/academies/20200430212059287c539aa14c5b64cb588d3744b5cc63.png "Into 2-D")

#### Ada beberapa kernel yang populer yaitu:

- Linear
- Polinomial
- RBF
- Sigmoid

### SVM untuk klasifikasi multi kelas

- menggunakan suatu teknik yaitu “One-vs-rest”
- SVM melakukan klasifikasi biner untuk masing-masing kelas.
- Kemudian membangun pemisah antara kelas ayam dan kelas bukan ayam, lalu pemisah antara kelas burger dan bukan kelas burger.

![](https://dicodingacademy.blob.core.windows.net/academies/2020043021222079d9cede7dc7887943d66af77fe4ea86.png "biner for one class")

### Latihan SKLearn SVM untuk classification

```python
import pandas as pd
df = pd.read_csv('sample_data/diabetes.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# memisahkan atribut pada dataset dan menyimpannya pada sebuah variabel
X = df[df.columns[:8]]
 
# memisahkan label pada dataset dan menyimpannya pada sebuah variabel
y = df['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler
 
 
# standarisasi nilai-nilai dari dataset
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
# memisahkan data untuk training dan testing
from sklearn.model_selection import train_test_split
 
 
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
# create SVM
from sklearn.svm import SVC
 
 
clf = SVC()
clf.fit(X_train, y_train)
```

```python
# membuat prediksi
clf.score(X_test, y_test)
```

## Support vector regression

- SVR menggunakan prinsip yang sama dengan SVM pada kasus klasifikasi.
- SVM berusaha mencari ‘jalan’ terbesar yang bisa memisahkan sampel-sampel dari kelas berbeda sedangkan
- SVR berusaha mencari jalan yang dapat menampung seluruh sampel.

### SVR untuk klasifikasi non-linier

- menggunakan trik kernel, yaitu mengubah data ke dalam dimensi yang lebih tinggi sehingga data dapat dipisahkan secara linier.

![](https://dicodingacademy.blob.core.windows.net/academies/202004302212029aed82c53d93feb0e117d0fe3558a996.png "SVR Non Linear")

### Latihan SKLearn SVR

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
 
 
data = pd.read_csv('sample_data/Salary_Data.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
import numpy as np
 
 
X = data['YearsExperience']
y = data['Salary']

# ketika hanya terdapat satu atribut pada dataframe, atribut perlu dirubah dgn numpy
X = X[:,np.newaxis]
```

```python
from sklearn.svm import SVR
model  = SVR(C=1000, gamma=0.05, kernel='rbf')
model.fit(X,y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.plot(X, model.predict(X))
```
